**Emperical rule**

mean-1*Std to mean+1std : 68%

mean-2*Std to mean+2std : 95%

mean-3*Std to mean+3std : 99.7%

step-1 : calculate mean of prevaialing wage

step-2 : calculate stf of prevaialing wage

step-3 : lb=mean-1*std

step-4 : ub=mean+1*std

step-5 : con1=data<ub

step-6 : con2=data<lb

step-7 : com3=con1 and con2

step-8 : data[con3]

step-9 : compair with 68% 25480 count

```python
wage_df=visa_df['prevailing_wage']
mean_wage=visa_df['prevailing_wage'].mean()
std_wage=visa_df['prevailing_wage'].std()
lb=mean_wage-(1*std)
ub=mean_wage+(1*std)
# con1=wage_df<ub
# con2=wage_df>lb
# con3=con1 & con2
# visa_df[con3]
visa_df[(wage_df>lb)&(wage_df<ub)]
```

```python
wage_df=visa_df['prevailing_wage']
mean_wage=visa_df['prevailing_wage'].mean()
std_wage=visa_df['prevailing_wage'].std()
lb=mean_wage-(2*std)
ub=mean_wage+(2*std)
visa_df[(wage_df>lb)&(wage_df<ub)]
```
