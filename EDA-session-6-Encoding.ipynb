**Map**

- mapping is one method to convert categorical values to numerical

- take one categorical column

- get the unique lables first

- make a dictonary with assign a number to each label

- for example case_status has 2 label

    - certified
    
    - denied

- assign 0 to certified and 1 to denied

- create a dictonary labels as keys and numbers as values only

- {'certified':0,'denied':1} 

```python
visa_df=pd.read_csv('Visadataset.csv')
for j in cat_col:
    dict1={}
    for k,i in enumerate(visa_df[j].unique(),0):
        dict1[i]=k
    dict1
    visa_df[j]=visa_df[j].map(dict1)
visa_df
```

```python
import math
x=visa_df['no_of_employees']
y=visa_df['yr_of_estab']
add1=0
add2=0
for i,j in zip(x,y):
    add1+=i
    add2+=j
numerator=add1*add2
x_var=x.var()
y_var=y.var()
denom=math.sqrt(x_var)*math.sqrt(y_var)
co_rel=numerator/denom
co_rel
```

```python
visa_df=pd.read_csv('Visadataset.csv')
# step-1 : read the package

from sklearn.preprocessing import LabelEncoder

# step-2 : save the package

le=LabelEncoder()

# step-3 : apply fit transform on data

visa_df['case_status']=le.fit_transform(visa_df['case_status'])
visa_df
```

**np.where**

- using np.where also we can change categorical data to numerical data

- but np.where is only for binary conditins

- np.where is same as if-else

- if condition is True then we apply true value

- if condition is false we apply false value

- so np.where use for only binary labels

- a column should have only unique label

```python
visa_df=pd.read_csv('Visadataset.csv')
visa_df['case_status']=np.where(visa_df['case_status']=='Certified',1,0)
visa_df
```

**Advantages**

- using one hot encoder we will get new columns

- the new column value at a time only one will be **true** another will be **false**

- which means new columns are **independent to each other**

- independent means no relation

- means no **multi-collineraity**

- multi-collineraity is the important property in machine learning

- **machine learning models expect the input columns should be independant to each other**

- do one hot encoder gives no multi-collineraity

**Disadvantages**

- the disadvantage is it will increase many columns

- in the above example the total columns becomes 25510

- this will happen because of case_id

- in real time we will drop the id columns

- even though we will drop we are getting 30 columns

- more columns means more time to process

- we required more infrastructure(CPU,GPU,TPU)

- this is called as **curse of dimensionality**
