**Take**

- first argument : indices

- indices will take some values in array which means in the format of list

- it will try to extract the data

- that values we provided presents rows or column?

- Axis=0 represents rows

- Axis=1 represents columns

- by default axis is mentioned as 0

```python
# Case-1
visa_df.take([1,2,3,4],axis=1)
```

```python
# Case-2
visa_df.take([100,200,300,400])
```

```python
visa_df.take([100,200,300,400],axis=1)
# now this will check for 100th,200th,300th and 400th column 
# but originally we have only 12 column(0,11)
# we will get error
```

```python
# Case-3 : select 3 columns 2nd 5th and 7th columns using take 
visa_df.take([2,5,7],axis=1)
```

```python
visa_df[2:9:3]
```
