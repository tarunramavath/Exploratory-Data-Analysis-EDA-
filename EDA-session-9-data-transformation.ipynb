**Log Transformation**

The natural logarithm `log` is the inverse of the exponential function,
so that `log(exp(x)) = x`. The natural logarithm is logarithm in base
`e`.

```python
x=10
np.log(x)
```

```python
log_data=np.log(exp_data)
plt.figure(figsize=(8,5))
plt.subplot(1,2,1).hist(log_data)
plt.subplot(1,2,2).hist(exp_data,bins=20,label='exp_data')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(exp_data,bins=20,label='exp_data')
plt.hist(log_data)
plt.show()
```

# **power transformer**

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method='box-cox')
visa_df['power_transformation_box_cox']=pt.fit_transform(visa_df[['prevailing_wage']])
visa_df[['prevailing_wage','power_transformation_box_cox']]
plt.hist(visa_df['power_transformation_box_cox'],bins=20)
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer()
visa_df['power_transformation_yeo_johnson']=pt.fit_transform(visa_df[['prevailing_wage']])
visa_df[['prevailing_wage','power_transformation_yeo_johnson']]
plt.hist(visa_df['power_transformation_yeo_johnson'],bins=20)
plt.show()
```

```python
visa_df[['prevailing_wage','power_transformation_box_cox','power_transformation_yeo_johnson']]
```

```python
plt.figure(figsize=(14,5))
plt.subplot(1,2,1).hist(visa_df['prevailing_wage'],bins=40)
plt.subplot(1,2,2).hist(visa_df['power_transformation_yeo_johnson'],bins=40)
plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.subplot(1,2,1).boxplot(data=visa_df,x='prevailing_wage',vert=False)
plt.subplot(1,2,2).boxplot(data=visa_df,x='power_transformation_yeo_johnson',vert=False)
plt.show()
```
