- create 3 lists using list comprehension

- list1= a value from 1 to 10

```python
import pandas as pd
```

```python
list1=[i for i in range(1,11)]
list2=[i*i for i in range(1,11)]
list3=[i*i*i for i in range(1,11)]
col_names=['numbers','Square','Cube']
df1=pd.DataFrame(zip(list1,list2,list3),columns=col_names)
df1
```

**Step-7 : how to drop a column**

- if ypu want to drop a specific column

- we will use drop method

- we alredy know anout difference between keyword and methods

- keywords will use directly

- methods will use with corresponding datafram

- the method name is drop

```python
# how to check the number of cloumns
DF.columns
```

```python
DF.drop
```

- column : which column you want to drop

- axis : represents the specified values is a row or column

  - axis=0 : means rows
 
  - axis=1 : means columns
 
  - by default axis=0

- inplace : will procide overwrite dataframe or not

  - this indicates overwrite the dataframe
 
  - inplace=False : this indicates the modified dataframe will be use when you provide new variable for the operation  

```python
DF.drop('number',axis=1) # dropes number column
DF.drop(10,axis=0) # dropes row 10
```

```python
DF # we can see original dataframe because we did not provided inplace in drop method
```
