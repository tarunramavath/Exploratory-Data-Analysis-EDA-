**Task**

- read the data

- develop a normalization data using logic

- step-1 : read the pwage data

- step-2 : min value

- step-3 : max value

- step-4 : nr : step-1 - step-2

- step-5 : dr : step-3 - step-2

- step-6 : nr/dr

```python
visa_df=pd.read_csv('Visadataset.csv')
wage_df=visa_df['prevailing_wage']
pwage_min=wage_df.min()
pwage_max=wage_df.max()
nr=wage_df-pwage_min
dr=pwage_max-pwage_min
res=nr/dr
visa_df['min_max_scalar']=res
visa_df
```
